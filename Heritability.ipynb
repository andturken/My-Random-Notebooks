```python
def rnorm(n,u,s):
    return np.random.standard_normal(n)*s+u
```

```python
# cause1: causal genetic factors shared by phenotype 1 and 2
# cause2: causal genetic factors exclusive to phenotype 1
N=1000;
cause1=rnorm(N,0,1);
cause2=rnorm(N,0,1);
cause3=rnorm(N,0,1);

h2=0.5; # heritability (A=h2, C=0, E=1-A)

alpha=1; # genetic relationship
Ga=alpha*cause1;
Gb=(1-alpha)*cause2;
Gc=(1-alpha)*cause3;
G1=(Ga+Gb)/np.sqrt(np.var(Ga)+np.var(Gb));
G2=(Ga+Gc)/np.sqrt(np.var(Ga)+np.var(Gc));
Phenotype_Sub1=np.sqrt(h2)*G1+np.sqrt(1-h2)*rnorm(N,0,1);
Phenotype_Sub2=np.sqrt(h2)*G2+np.sqrt(1-h2)*rnorm(N,0,1);
r1=np.corrcoef(Phenotype_Sub1,Phenotype_Sub2)[0,1];

alpha=0.5; # genetic relationship
Ga=alpha*cause1;
Gb=(1-alpha)*cause2;
Gc=(1-alpha)*cause3;
G1=(Ga+Gb)/np.sqrt(np.var(Ga)+np.var(Gb));
G2=(Ga+Gc)/np.sqrt(np.var(Ga)+np.var(Gc));
Phenotype_Sub1=np.sqrt(h2)*G1+np.sqrt(1-h2)*rnorm(N,0,1);
Phenotype_Sub2=np.sqrt(h2)*G2+np.sqrt(1-h2)*rnorm(N,0,1);
r2=np.corrcoef(Phenotype_Sub1,Phenotype_Sub2)[0,1];

Falconer_A=2*(r1-r2);
Falconer_C=r1-Falconer_A;
Falconer_E=1-Falconer_A-Falconer_C;
print Falconer_A,Falconer_C,Falconer_E;
```
