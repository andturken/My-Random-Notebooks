```python
%pylab inline

import numpy as np
import scipy.linalg as lin
import scipy.stats as sst
import pylab
```

```python
def rnorm(n,u,s):
    return np.random.standard_normal(n)*s+u
```

```python
N=1000;

cause1=rnorm(N,0,1);
cause2=rnorm(N,0,1);
cause3=rnorm(N,0,1);
cause4=rnorm(N,0,1);
cause5=rnorm(N,0,1);
cause6=rnorm(N,0,1);
```

```python
pylab.hist(cause1,30);
```

```python
pylab.hist(cause2,30);
```

```python
pylab.scatter(cause1,cause2,alpha=0.1);
```

```python
h2=0.80;

G=(cause1+cause2)/np.sqrt(2);
Phenotype1_MZ_Twin1=np.sqrt(h2)*G+np.sqrt(1-h2)*rnorm(N,0,1);
Phenotype1_MZ_Twin2=np.sqrt(h2)*G+np.sqrt(1-h2)*rnorm(N,0,1);

G1=(cause3+cause4)/np.sqrt(2);
G2=(cause3+cause5)/np.sqrt(2);
Phenotype1_DZ_Twin1=np.sqrt(h2)*G1+np.sqrt(1-h2)*rnorm(N,0,1);
Phenotype1_DZ_Twin2=np.sqrt(h2)*G2+np.sqrt(1-h2)*rnorm(N,0,1);
```

```python
pylab.scatter(Phenotype1_MZ_Twin1,Phenotype1_MZ_Twin2,alpha=0.1);
```

```python
pylab.scatter(Phenotype1_DZ_Twin1,Phenotype1_DZ_Twin2,alpha=0.1);
```

```python
rMZ=np.corrcoef(Phenotype1_MZ_Twin1,Phenotype1_MZ_Twin2)[0,1];
rDZ=np.corrcoef(Phenotype1_DZ_Twin1,Phenotype1_DZ_Twin2)[0,1];
print rMZ,rDZ
```

```python
Falconer_heritability=2*(rMZ-rDZ);
print Falconer_heritability
```
