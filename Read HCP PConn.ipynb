Nifti 2

```python
%matplotlib inline
import numpy as np
import pylab as pl
from scipy import stats
import struct;
import xml.etree.ElementTree as ET
from StringIO import StringIO
import base64 as b64
import zlib

dir="/Users/roberto/Documents/_04_Data/2014_10HCP/HCP500_Parcellation_Timeseries_Netmats/netmats/";
f=open(dir+"3T_Q1-Q6related468_MSMsulc_d25_ts2_netmat1/100307.pconn.nii","r");

# Read header
f.seek(0)
c=f.read(540)
hdr=struct.unpack("i8shh8qddd8dqddddddqq80s24siidddddd4d4d4diii16sc15s",c);
print \
"sizeof_hdr",hdr[0],"magic",hdr[1],"datatype",hdr[2],"bitpix",hdr[3],"dim",hdr[4:12],\
"intent_p1",hdr[12],"intent_p2",hdr[13], "intent_p3",hdr[14],"pixdim",hdr[15:23],\
"vox_offset",hdr[23],"scl_slope",hdr[24],"scl_inter",hdr[25],"cal_max",hdr[26],\
"cal_min",hdr[27],"slice_duration",hdr[28],"toffset",hdr[29],"slice_start",hdr[30],\
"slice_end",hdr[31],"descrip",hdr[32],"aux_file",hdr[33],"slice_code",hdr[54],\
"xyzt_units",hdr[55],"intent_code",hdr[56],"intent_name",hdr[57]

# Read connectivity matrix
f.seek(hdr[23]);       # Move to where the matrix data begins
N=hdr[9];              # Matrix size: number of parcels
d=f.read(N*N*4);       # Read the matrix
f.seek(540);           # Read XML

# Read CIFTI xml description
'''
extension is 0000 if there's no extension; 1000 if there's one
esize is the number of bytes that form the extended header data
       + esize must be a positive integral multiple of 16
       + this length includes the 8 bytes of esize and ecode themselves
ecode is a non-negative integer that indicates the format of the
       extended header data that follows (32 for CIFTI)
'''
ext=s.unpack("4s2i",f.read(12));
print "extension, esize, ecode",ext;
esize=ext[1];
f.seek(540+12);
str=f.read(esize-8-13); # There were several 0s that prevented ET from parsing str
root = ET.fromstring(str) # https://docs.python.org/2/library/xml.etree.elementtree.html
f.close();

# Connectivity
#--------------
# Plot connectivity matrix
M=s.unpack("%df"%(N*N),d);
R=np.reshape(M,[N,N])
pl.matshow(R, cmap=pl.cm.RdBu_r)   # Plot the matrix

# Parcels
#---------

# Get transformation matrix for voxel parcels
T=np.reshape(np.loadtxt(StringIO(root.find("./Matrix/MatrixIndicesMap/Volume/TransformationMatrixVoxelIndicesIJKtoXYZ").text)),[4,4]);
print T;

# Read left hemisphere's surface
tree=ET.parse("/Users/roberto/Documents/_04_Data/2014_10HCP/HCP_WB_Tutorial_Beta0.83/249947/MNINonLinear/fsaverage_LR32k/249947.L.pial.32k_fs_LR.surf.gii");
gifti=tree.getroot()
gz=b64.b64decode(gifti.find("./DataArray/Data").text)
Ldata=zlib.decompress(gz)
Lverts=np.reshape(struct.unpack("%if"%(len(Ldata)/4),Ldata),[len(Ldata)/4/3,3])

# Read right hemisphere's surface
tree=ET.parse("/Users/roberto/Documents/_04_Data/2014_10HCP/HCP_WB_Tutorial_Beta0.83/249947/MNINonLinear/fsaverage_LR32k/249947.R.pial.32k_fs_LR.surf.gii");
gifti=tree.getroot()
gz=b64.b64decode(gifti.find("./DataArray/Data").text)
Rdata=zlib.decompress(gz)
Rverts=np.reshape(struct.unpack("%if"%(len(Rdata)/4),Rdata),[len(Rdata)/4/3,3])

#print np.mean(Lverts,axis=0)    # this one should be to the left
#print np.mean(Rverts,axis=0)    # this one should be to the right
```

```python
# Find parcel's barycentres in voxel and surface space
ba=[];
for parcel in root.iter("Parcel"):
    mcoord=np.array([0,0,0]);
    n=0;

    # Voxels
    vijk=parcel.find("VoxelIndicesIJK");
    if vijk is not None:
        coords=np.loadtxt(StringIO(vijk.text));
        if(len(coords.shape)==2):
            for i in coords:
                i=np.append(i,[1],axis=0);
                mcoord=mcoord+T.dot(np.transpose(i))[0:3];
                n=n+1;
        else:
            i=np.append(coords,[1],axis=0);
            mcoord=mcoord+T.dot(np.transpose(i))[0:3]
            n=n+1

    # Left surface
    vsurf=parcel.find("./Vertices[@BrainStructure='CIFTI_STRUCTURE_CORTEX_LEFT']");
    if vsurf is not None:
        ind=np.loadtxt(StringIO(vsurf.text));
        if(len(ind.shape)>0):
            for i in range(1,len(ind)):
                mcoord=mcoord+Lverts[ind[i]];
                n=n+1;

    # Right surface
    vsurf=parcel.find("./Vertices[@BrainStructure='CIFTI_STRUCTURE_CORTEX_RIGHT']");
    if vsurf is not None:
        ind=np.loadtxt(StringIO(vsurf.text));
        if(len(ind.shape)>0):
            mcoord=np.array([0,0,0]);
            for i in range(1,len(ind)):
                mcoord=mcoord+Rverts[ind[i]];
                n=n+1;

    # Append to barycenter list
    ba=np.append(ba,mcoord/n)

ba=ba.reshape([len(ba)/3,3]);

# Draw the parcel's barycentres
from mpl_toolkits.mplot3d import Axes3D
fig = pl.figure()
ax = fig.add_subplot(111, projection='3d')
pl.scatter(ba[:,0],ba[:,1],zs=ba[:,2],zdir=u'z', s=20, c=u'b')
```

```python
# Write an analyze volume with the voxel and surface parcels
vx=[0]*(91*109*91);
iparcel=0;
for parcel in root.iter("Parcel"):
    iparcel=iparcel+1;

    # Voxels
    vijk=parcel.find("VoxelIndicesIJK");
    if vijk is not None:
        coords=np.loadtxt(StringIO(vijk.text));
        if(len(coords.shape)==2):
            for i in coords:
                vx[int(i[2])*109*91+int(i[1])*91+int(i[0])]=iparcel;
        else:
            i=np.append(coords,[1],axis=0);
            vx[i[2]*109*91+i[1]*91+i[0]]=iparcel;

    # Left surface
    vsurf=parcel.find("./Vertices[@BrainStructure='CIFTI_STRUCTURE_CORTEX_LEFT']");
    if vsurf is not None:
        ind=np.loadtxt(StringIO(vsurf.text));
        if(len(ind.shape)>0):
            for i in range(1,len(ind)):
                j=np.linalg.inv(T).dot(np.transpose(np.append(Lverts[ind[i]],[1])));
                vx[int(j[2])*109*91+int(j[1])*91+int(j[0])]=iparcel;

    # Right surface
    vsurf=parcel.find("./Vertices[@BrainStructure='CIFTI_STRUCTURE_CORTEX_RIGHT']");
    if vsurf is not None:
        ind=np.loadtxt(StringIO(vsurf.text));
        if(len(ind.shape)>0):
            mcoord=np.array([0,0,0]);
            for i in range(1,len(ind)):
                j=np.linalg.inv(T).dot(np.transpose(np.append(Rverts[ind[i]],[1])));
                vx[int(j[2])*109*91+int(j[1])*91+int(j[0])]=iparcel;

fimg=open("/Users/roberto/Desktop/parcels.img","wb");
buf=struct.pack("%ih"%(91*109*91),*vx);
fimg.write(buf);
fimg.close();
```
