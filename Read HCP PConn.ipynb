```python
%matplotlib inline
import numpy as np
import pylab as pl
from   scipy import stats
import struct;
import xml.etree.ElementTree as ET
from   StringIO import StringIO
import base64 as b64
import zlib
```

```python
# Read .pconn.nii file
def read_pconn(filename):

    result={}; # empty dictionary

    f=open(filename,"r");

    # 1. Read header
    c=f.read(540)
    hdr=struct.unpack("i8shh8qddd8dqddddddqq80s24siidddddd4d4d4diii16sc15s",c);
    header={"sizeof_hdr":hdr[0],"magic":hdr[1],"datatype":hdr[2],"bitpix":hdr[3],"dim":hdr[4:12],\
    "intent_p1":hdr[12],"intent_p2":hdr[13], "intent_p3":hdr[14],"pixdim":hdr[15:23],\
    "vox_offset":hdr[23],"scl_slope":hdr[24],"scl_inter":hdr[25],"cal_max":hdr[26],\
    "cal_min":hdr[27],"slice_duration":hdr[28],"toffset":hdr[29],"slice_start":hdr[30],\
    "slice_end":hdr[31],"descrip":hdr[32],"aux_file":hdr[33],"slice_code":hdr[54],\
    "xyzt_units":hdr[55],"intent_code":hdr[56],"intent_name":hdr[57]};
    result["hdr"]=header;

    # 2. Read extension
    '''
    extension is 0000 if there's no extension; 1000 if there's one
    esize is the number of bytes that form the extended header data
           + esize must be a positive integral multiple of 16
           + this length includes the 8 bytes of esize and ecode themselves
    ecode is a non-negative integer that indicates the format of the
           extended header data that follows (32 for CIFTI)
    '''
    f.seek(540);           
    ext=struct.unpack("4s2i",f.read(12));
    extension={"extension":ext[0],"esize":ext[1],"ecode":ext[2]}
    result["ext"]=extension;
    #print "extension, esize, ecode",ext;
    
    # 3. Read CIFTI xml description
    f.seek(540+12);
    str=f.read(extension["esize"]-8-13); # There were several 0s that prevented ET from parsing str
    root = ET.fromstring(str) # https://docs.python.org/2/library/xml.etree.elementtree.html
    result["xml"]=root;
    
    # 4. Read connectivity matrix
    f.seek(hdr[23]);           # Move to where the matrix data begins
    N=hdr[9];                  # Matrix size: number of parcels
    d=f.read(N*N*4);           # Read the matrix
    M=struct.unpack("%df"%(N*N),d); # Unpack connectivity data
    R=np.reshape(M,[N,N])
    result["n"]=N
    result["matrix"]=R;
    
    f.close();
    
    return result;
```

```python
# Read .gii files
def read_gii(filename):
    
    result={}
    
    # Read xml header
    gifti=ET.parse(filename).getroot();
    result["xml"]=gifti;
    
    # Read vertices data
    gz=b64.b64decode(gifti.find("./DataArray/Data").text)
    data=zlib.decompress(gz)
    verts=np.reshape(struct.unpack("%if"%(len(data)/4),data),[len(data)/4/3,3])
    result["verts"]=verts;
    
    return result
```

```python
# Read .dtseries.nii file
def read_dtseries(filename):

    result={}; # empty dictionary

    f=open(filename,"r");

    # 1. Read header
    c=f.read(540)
    hdr=struct.unpack("i8shh8qddd8dqddddddqq80s24siidddddd4d4d4diii16sc15s",c);
    header={"sizeof_hdr":hdr[0],"magic":hdr[1],"datatype":hdr[2],"bitpix":hdr[3],"dim":hdr[4:12],\
    "intent_p1":hdr[12],"intent_p2":hdr[13], "intent_p3":hdr[14],"pixdim":hdr[15:23],\
    "vox_offset":hdr[23],"scl_slope":hdr[24],"scl_inter":hdr[25],"cal_max":hdr[26],\
    "cal_min":hdr[27],"slice_duration":hdr[28],"toffset":hdr[29],"slice_start":hdr[30],\
    "slice_end":hdr[31],"descrip":hdr[32],"aux_file":hdr[33],"slice_code":hdr[54],\
    "xyzt_units":hdr[55],"intent_code":hdr[56],"intent_name":hdr[57]};
    result["hdr"]=header;

    # 2. Read extension
    # extension is 0000 if there's no extension; 1000 if there's one
    # esize is the number of bytes that form the extended header data
    #        + esize must be a positive integral multiple of 16
    #        + this length includes the 8 bytes of esize and ecode themselves
    # ecode is a non-negative integer that indicates the format of the
    #        extended header data that follows (32 for CIFTI)
    f.seek(540);           
    ext=struct.unpack("4s2i",f.read(12));
    extension={"extension":ext[0],"esize":ext[1],"ecode":ext[2]}
    result["ext"]=extension;
    #print "extension, esize, ecode",ext;
    
    # 3. Read CIFTI xml description
    f.seek(540+12);
    str=f.read(extension["esize"]-12-2); # There were several 0s that prevented ET from parsing str
    #print(str)
    root = ET.fromstring(str) # https://docs.python.org/2/library/xml.etree.elementtree.html
    result["xml"]=root;
        
    # 4. Read connectivity matrix
    f.seek(hdr[23]);           # Move to where the matrix data begins
    nsamples=hdr[9];           # number of time samples per series
    nlocs=hdr[10];             # number of locations (vertices + voxels)

    d=f.read(nsamples*nlocs*4);         # Read the matrix
    D=struct.unpack("%df"%(nsamples*nlocs),d); # Unpack connectivity data
    B=np.reshape(D,[nsamples,nlocs])
    result["nsamples"]=nsamples;
    result["nlocs"]=nlocs;
    result["matrix"]=B;
    
    f.close();
    
    return result;
```
