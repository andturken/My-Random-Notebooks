```python
%matplotlib inline
import os, sys
import shutil
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# BigBrain slice size
W=125826
H=86165

# Chunk size
d=16384

# Chunk dzi directory
dir="/Volumes/cinq/rto/data/bigbrain/dzi/"

# Final image
final="/Library/WebServer/Documents/virtualmicroscope/final/"

# Reconstituted image size
bim=Image.new("L",(256*(W/d+1),256*(H/d+1)))
```

```python
# Reconstitute image
c=0;
for i in range(0,W,d):
    r=0;
    for j in range(0,H,d):
        filename="bb."+str(i)+"."+str(j)+"_files/";
        tmp=Image.open(dir+filename+"8/0_0.jpg");
        bim.paste(tmp,(c*256,r*256));
        r+=1;
    c+=1;
```

```python
plt.imshow(np.asarray(bim))
```

```python
# Crop to appropriate size
dx=256-tmp.size[0]
dy=256-tmp.size[1]
bim2=bim.crop((0,0,bim.size[0]-dx,bim.size[1]-dy))
plt.imshow(np.asarray(bim2))
```

```python
# Save for test
bim2.save("/Users/roberto/Desktop/test.png")
```

```python
# Write upward tiles
for k in range(1,12):
    dir1=final+"final_files/"+str(11-k)+"/";
    print(dir1);
    if not os.path.exists(dir1):
        os.makedirs(dir1)
    g=2**k;
    A=256*g;
    c=0;
    for i in range(0,bim2.size[0],A):
        r=0;
        for j in range(0,bim2.size[1],A):
            sz=(min(A,bim2.size[0]-i),min(A,bim2.size[1]-j));
            cr=bim2.crop((i,j,i+sz[0],j+sz[1]));
            sz1=((sz[0]+g-1)/g,(sz[1]+g-1)/g);
            cr=cr.resize(sz1,Image.ANTIALIAS);
            cr.save(dir1+str(c)+"_"+str(r)+".jpg");
            print(k,c,r,i,j,sz[0],sz[1],sz1[0],sz1[1],str(c)+"_"+str(r)+".jpg");
            #plt.figure();
            #plt.imshow(np.asarray(cr));
            r+=1;
        c+=1;
```

```python
# Rename chunk's tiles for final dzi
for level in range(0,7):
    if not os.path.exists(final+"final_files/"+str(level+11)):
        os.makedirs(final+"final_files/"+str(level+11));
    n=0;
    g=2**level;
    c=0;
    for i in range(0,W,d):
        r=0;
        for j in range(0,H,d):
            for k in range(0,g):
                for l in range(0,g):
                    src=dir+"bb."+str(i)+"."+str(j)+"_files/"+str(level+8)+"/"+str(k)+"_"+str(l)+".jpg";
                    dst=final+"final_files/"+str(level+11)+"/"+str(r*g+k)+"_"+str(c*g+l)+".jpg"
                    if(os.path.isfile(src)):
                        shutil.copyfile(src,dst)
                        n+=1;
            r+=1;
        c+=1;
    print(n,"images for level",level+11)
```
