#We all are from the same big family

```python
%matplotlib inline
import numpy as np;
import matplotlib.pyplot as plt;

import scipy
import scipy.cluster.hierarchy as sch
from scipy import linalg, stats
from sklearn.decomposition import PCA

def rnorm(n,u,s):
    return np.random.standard_normal(n)*s+u

def imshowcluster(D):
    # Based on http://stackoverflow.com/questions/2982929/plotting-results-of-hierarchical-clustering-ontop-of-a-matrix-of-data-in-python
    fig = plt.figure()

    # Compute and plot dendrogram.
    axdendro1 = fig.add_axes([0.0,0.2,0.19,0.8])
    Y1 = sch.linkage(D, method='centroid')
    Z1 = sch.dendrogram(Y1, orientation='right')
    axdendro1.set_xticks([])
    axdendro1.set_yticks([])

    # Compute and plot dendrogram.
    axdendro2 = fig.add_axes([0.2,0.00,0.6,0.19])
    Y2 = sch.linkage(D.transpose(), method='centroid')
    Z2 = sch.dendrogram(Y2, orientation='bottom')
    axdendro2.set_xticks([])
    axdendro2.set_yticks([])

    # Plot distance matrix.
    axmatrix = fig.add_axes([0.2,0.2,0.6,0.8])
    idx1 = Z1['leaves']
    idx2 = Z2['leaves']
    D = D[idx1,:]
    D = D[:,idx2]
    im = axmatrix.matshow(D, aspect='auto', origin='lower')
    axmatrix.set_xticks([])
    axmatrix.set_yticks([])

    # Plot colorbar.
    axcolor = fig.add_axes([0.91,0.1,0.02,0.8])
    plt.colorbar(im, cax=axcolor)

    # Add axes labels
    axmatrix.set_xticks(range(D.shape[0]))
    axmatrix.set_xticklabels(idx1, minor=False)
    axmatrix.xaxis.set_label_position('bottom')
    axmatrix.xaxis.tick_bottom()

    plt.xticks(rotation=-90, fontsize=4)

    axmatrix.set_yticks(range(D.shape[1]))
    axmatrix.set_yticklabels(idx2, minor=False)
    axmatrix.yaxis.set_label_position('right')
    axmatrix.yaxis.tick_right()
    # Display and save figure.
    #fig.savefig('dendrogram.png')
```

#First approach
##Mixing ancestral genomes

## Section 1: Production of a population of related subjects

Suppose there are $Ne$ independent ancestors. Each ancestor $i$ has a genome that makes that for a given phenotype, its value is $Ze_i$. For simplicity, assume that Ze is exclusively produced by the ancestor's genome.

```python
Ne=100; # the non-observed original independent ancestors
Ze=rnorm(Ne,0,1); # the non-observed original phenotypes
```

The observed population consists of N subjects. The genome of each of this subjects is a random mix of the ancestral genomes. The phenotype of each subject is then a weighted mix of the ancestor's phenotypes. For simplicity, we suppose again that the phenotype is exclusively produced by the subject's genome.

```python
N=10000; # the observed population
G=np.zeros((N,Ne)); # their non-observed relative proportion of each of the ancestors
Z=np.zeros(N); # the genetic contribution to the observed phenotypes
```

The degree of genetic relatedness of a pair of subjects reflects how much their proportions of ancestral genome are similar. All degrees of genetic relatedness among subjects are stored in the kinship matrix $K$.

```python
K=np.zeros((N,N)); # their real kinship matrix, based on the non-observed proportions of independent ancestral genome
```

As we said, phenotypes are exclusively determined by the subject's genome. Their heritability is 1:

```python
h2=1; # phenotype heritability
P=np.zeros(N); # the observed phenotypes
```

First, we compute the phenotypes of the subjects in the population. First we generate mixing proportions vectors for each subject. We draw the proportion of ancestral genome $i$ from a uniform distribution. The total proportions are normalised so that they add to 1.

```python
# First part: Make N subjects by randomly mixing the Ne independent genomes
# 1: make a vector with the mixing proportions, the genomes
for i in range(0,N):
    prop=np.random.uniform(0,1,Ne);
    G[i,:]=prop/np.sum(prop);
```

Second, we compute the kinship matrix by comparing the similarity of the ancestral genome proportions of each pair of subjects. For each ancestral genome (among the $Ne$ possible) we compute the absolute difference in proportion. The minimum value is 0, in the case both subjects have the same proportion of that ancestral genome. The maximum value is 1, in case one subjects inherits all her genome from an ancestor, whereas the second subjects does not inherit any genome from that same ancestor. When all proportions of ancestral genomes are taken together, the minimum value of the difference is 0 (identical genomes); the maximum value is 2, in the extreme case where one subject inherits all her genome from one ancestor and the other subject inherits all her genome from a different ancestor.
The degree of relatedness is then

$K_{ij}=1-\frac {\sum |G_i-Gj|} 2$,

which takes value from 0 (completely different) to 1 (identical genomes).

```python
# 2: compute the kinship matrix K
for i in range(0,N):
    for j in range(i,N):
        K[i,j]=1-np.sum(np.abs(G[i]-G[j]))/2.0;
        K[j,i]=K[i,j];
```

Using each subject's genome proportions, compute their phenotype as

$Z=G \cdot Z_e$.

Each row $i$ of $G$ represent the ancestral genome proportions of subject $i$, and $Z_e$ is the phenotype associated to each of the ancestral genomes. The population phenotypes $Z$ are normalised to 0 mean and standard deviation of 1.

```python
# 3: compute the genomic influence on the phenotype
Z=G.dot(Ze);
Z=(Z-Z.mean())/Z.std(); # normalised
```

We have been using $h^2=1$, but if we wanted to add some individual randomness, and just by changing the value of $h^2$ at the beginning, the following code would produce a phenotype with the specified heritability (mean centered and normalised):

```python
# 4: add individual randomness to produce the observed phenotype with heritability h2
P=Z*np.sqrt(h2)+rnorm(N,0,1)*np.sqrt(1-h2);
P=(P-P.mean())/P.std(); # normalised
```

## Section 2: Estimation of $h^2$ from a population of related subjects

The first part of the notebook was concerned with the production of a population of subjects with varying degrees of genetic relationship. This second part aims at using that population to compute $h^2$.

From Visscher et al (2014): "For unrelated individuals, where the phenotypic correlation between individuals is small, mixed linear model analysis using the REML approach is asymptotically equivalent to simple regression analysis of pairwise phenotypic similarity/difference on pairwise genetic similarity, as measured by identity-by-descent (IBD) or identity-by-state (IBS) at genome-wide markers [17â€“20]. Under such circumstance, a regression of the cross-product of the phenotypes is equivalent to using both the squared difference and squared sum of the pairwise phenotypes, and using the cross- product is equivalent to using maximum likelihood [19]. The model for the regression-based analysis can be written as

$z_{ij}=\mu+b A_{ij} + \epsilon_{ij}$"

Here $z_{ij}=y_i y_j$, the product of the phenotypes, $A_{ij}$ is the $ij$-th element of the genetic relationship matrix, and $\epsilon_{ij}$ is the residual of the regression. The coefficient $b$ is equivalent to $\sigma_G^2$.

```python
# Compute h2 using a subsample of all the pairs of subjects
M=10000;
zz=np.zeros(M);
a=np.zeros(M);
k=0;
for k in range(M):
    i=int(N*np.random.rand());
    j=int(N*np.random.rand());
    zz[k]=P[i]*P[j];
    a[k]=K[i,j];
slope, intercept, r_value, p_value, std_err = stats.linregress(a,zz);
h2=slope/np.var(P);

print "slope:",slope;
print "intercept:",intercept;
print "correlation:",r_value;
print "p_value:",p_value;
print "std_err:",std_err;
print "h2:",h2;
```

#Second approach
##Unrelated individuals as correlated variables
The observed subjects result from a combination of a smaller number of uncorrelated ancestors


```python
N=10000; # observed subjects
Ne=100; # unobserved independent ancestors
rng=np.random.RandomState(123); # init random seed

# Draw random phenotypic values for each ancestor
Pe=rng.randn(Ne);

# Create N observed subjects as random combinations of Ne uncorrelated ancestors
G=rng.randn(N,Ne);
for i in range(N):
    G[i,:]/=linalg.norm(G[i,:]);

# Compute genetic part of the N observed phenotypes
# and add noise to produce a target heritability h2t
Pg=G.dot(Pe);
h2t=0.8;
P=h2t*Pg/np.var(Pg)+(1-h2t)*rng.randn(N);

# Kinship matrix
K=np.corrcoef(G);

# Compute h2 using a subsample of all the pairs of subjects
M=10000;
zz=np.zeros(M);
a=np.zeros(M);
k=0;
for k in range(M):
    i=int(N*np.random.rand());
    j=int(N*np.random.rand());
    zz[k]=P[i]*P[j];
    a[k]=K[i,j];
slope, intercept, r_value, p_value, std_err = stats.linregress(a,zz);
h2=slope/np.var(P);

print "slope:",slope;
print "intercept:",intercept;
print "correlation:",r_value;
print "p_value:",p_value;
print "std_err:",std_err;
print "h2:",h2;
```

##Generate monozygotic and dizygotic twins

```python
N=10000; # observed subjects
Ne=100; # unobserved independent ancestors
rng=np.random.RandomState(123); # init random seed

# Draw random phenotypic values for each ancestor
Pe=rng.randn(Ne);

# Allocate memory for the genomes
G=np.zeros((N,Ne));

# Create N/4 pairs of monozygotic twins
# MZ twins share the same combination of ancestral genome
Gmz=rng.randn(N/4,Ne);
for i in range(N/4):
    G[2*i,:]=Gmz[i,:];
    G[2*i+1,:]=Gmz[i,:];

# Create N/4 pairs of dizygotic twins
# DZ twins share half of the combination of ancestral genome
Gdz=rng.randn(N/2,Ne);
for i in range(0,N/2,2):
    Gdz[i,:(Ne/2)]=Gdz[i+1,:(Ne/2)]
G[N/2:,:]=Gdz;
    
# Normalize
for i in range(N):
    G[i,:]/=linalg.norm(G[i,:]);

# Compute genetic part of the N observed phenotypes
# and add noise to produce a target heritability h2t
Pg=G.dot(Pe);
h2t=0.8;
P=h2t*Pg/np.var(Pg)+(1-h2t)*rng.randn(N);

# Kinship matrix
K=np.corrcoef(G);

# Linear regression method using a subsample of all pairs of subjects
M=10000;
zz=np.zeros(M);
a=np.zeros(M);
k=0;
for k in range(M):
    i=int(N*np.random.rand());
    j=int(N*np.random.rand());
    zz[k]=P[i]*P[j];
    a[k]=K[i,j];
slope, intercept, r_value, p_value, std_err = stats.linregress(a,zz);
h2=slope/np.var(P);

print "slope:",slope;
print "intercept:",intercept;
print "correlation:",r_value;
print "p_value:",p_value;
print "std_err:",std_err;
print "h2:",h2;

# Falconer's method
pmz1=P[range(0,N/2,2)]
pmz2=P[range(1,N/2+1,2)]
rmz=np.corrcoef(pmz1,pmz2)[0,1];

pdz1=P[range(N/2,N,2)];
pdz2=P[range(N/2+1,N,2)];
rdz=np.corrcoef(pdz1,pdz2)[0,1];

print "h2:",2*(rmz-rdz);
```

##Correlated phenotypes in a population of unrelated individuals

```python
N=1000; # number of subjects in the observed population
Nr=9; # number of (correlated) phenotypes
Ni=3; # number of independent effects
rng = np.random.RandomState(123); # init random seed

# 1. Create (correlated) phenotypes
# Linear combination of Ni uncorrelated variables that produce the Nr correlated variables.
# (combination vectors are of unitary norm)
M=rng.randn(Nr,Ni);
for i in range(Nr):
    M[i,:]/=linalg.norm(M[i,:]);

# Scale with random eigenvalues
W=-np.sort(-np.abs(rng.randn(Ni)));
M=M.dot(np.diag(W));

# Random Ni values
Pi=rng.randn(N,Ni);

# Random Nr values produced by combining the independent Ni values
# The Nr regions are correlated among them
P=Pi.dot(M.T);
m=np.corrcoef(P.T);
plt.figure();
#plt.imshow(m,interpolation="nearest");
imshowcluster(m);

# 2. Try to guess M from P
U,S,VT=linalg.svd(P,full_matrices=False)
S=np.diag(S);
SVT=S.dot(VT);
M1=SVT[:Ni,:].T;

plt.figure();
plt.subplot(1,2,1);
plt.imshow(M,interpolation="nearest");
plt.subplot(1,2,2);
plt.imshow(M1,interpolation="nearest");

plt.figure();
plt.subplot(1,2,1);
plt.plot(M)
plt.subplot(1,2,2);
plt.plot(M1)

#X_homo = X + sigma * rng.randn(N, Nr); # Adding homoscedastic noise
#sigmas = sigma * rng.rand(Nr) + sigma / 2.0; # Adding heteroscedastic noise
#X_hetero = X + rng.randn(N, Nr) * sigmas
```
