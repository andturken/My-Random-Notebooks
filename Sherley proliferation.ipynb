# Sherley's equation for cell proliferation

Cell proliferation is often modelled by an exponential growth, however, because of asymmetric divisions with varying rates of dividing cells, the exponential model is not always accurate.

Sherley et al (1995) proposed a model taking into account asymmetric division and fraction of dividing cells. The model considers an original population of $N$ cells. When these cells divide, the daughters can include proliferating cells and dividing cells which will each produce 2 cells. The fraction of dividing cells is $F_d$. After one step of cell proliferation, de population will be then composed of $(1-F_d) N$ non-dividing cells and $2 F_d N$ cells resulting from cell division. The total number of cells will be then $(1-F_d)N + 2 F_d N$. In the following step, only the dividing cells will continue to proliferate, so that the total number of cells will be $(1-F_d)N + 2 (1-F_d) F_d N + (2 F_d)^2 N$. Sherley et al conclude that the series for the total number of cell after $k$ steps of cell proliferation is:

$N(k) = \frac N 2 \left ( 1 + \frac {1-(2 F_d)^{k+1}} {1-2 F_d} \right )$.

```python
%pylab inline
import numpy as np
from numpy import linalg as LA
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# N steps of cell proliferation
N=100;
Fd0=0.49;
Fd1=0.5000000001;
Fd2=0.51;


S=pd.DataFrame([[N,N,N]]);
for k in range(1,N):
    S.loc[k]=[
        (N/2)*(1+(1-(2*Fd0)**(k+1))/(1-2*Fd0)),
        (N/2)*(1+(1-(2*Fd1)**(k+1))/(1-2*Fd1)),
        (N/2)*(1+(1-(2*Fd2)**(k+1))/(1-2*Fd2))
    ];

# Plot
ax=S.plot()
ax.set_xlabel("Iteration");
ax.set_ylabel("Number of cells");

ax=S.plot(logy=True)
ax.set_xlabel("Iteration");
ax.set_ylabel("Log Number of cells");
```
