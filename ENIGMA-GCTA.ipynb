# Using liftOver to move everything to hg19



liftOver will just move rs numbers from hg18 to hg19, for example, but it will not do all the rest... UCSC people provide two other scripts for that: LiftMap.py and LiftRsNumber.py.
All the process and information is described here: http://genome.sph.umich.edu/wiki/LiftOver
There are two adviced methods to lift a complete ped/map plink dataset. The 1s uses liftMap.py and will only lift positions but not names (it lifts, however, most of the genome). The 2nd one uses LiftRsNumber.py and lifts both positions and names.
